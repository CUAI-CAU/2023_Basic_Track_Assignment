sklearn에서 제공하는 dataset(iris data)을 활용해 진행  
load_iris()를 쓰면, iris data를 불러옴  
type을 확인해보면 Bunch라고 나옴

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
print(type(iris_data))
```

load_iris로 불러온 bunch data는 keys() 메서드가 있음  
사용하면, 데이터셋의 key들을 반환함

```python
keys = iris_data.keys()
print('붓꽃 데이터 세트의 키들:', keys)
```

데이터셋은 feature_names라는 속성 존재  
-> len이 4인 list 자료형으로, 4개의 특성명을 담고 있음  
  
데이터셋은 또한 target_names 속성 존재  
prediction을 통해 맞추려는 label 명을 담고 있는 ndarray  
  
데이터셋은 data 속성 존재  
실제 data를 ndarray로 반환함  
(iris_data['data']로도 접근 가능)  
  
데이터셋은 target 속성 존재
data 별 target을 index 맞게 저장  
ndarray로 저장

```python
print('\n feature_names 의 type:',type(iris_data.feature_names))
print(' feature_names 의 shape:',len(iris_data.feature_names))
print(iris_data.feature_names)

print('\n target_names 의 type:',type(iris_data.target_names))
print(' target_names 의 shape:',len(iris_data.target_names))
print(iris_data.target_names)

print('\n data 의 type:',type(iris_data.data))
print(' data 의 shape:',iris_data.data.shape)
print(iris_data['data'])

print('\n target 의 type:',type(iris_data.target))
print(' target 의 shape:',iris_data.target.shape)
print(iris_data.target)
```
