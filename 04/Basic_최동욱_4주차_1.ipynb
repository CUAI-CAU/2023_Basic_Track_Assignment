### 분류  
학습 데이터로 주어진 데이터의 피처와 레이블값(결정 값, 클래스 값)을 머신러닝 알고리즘으로 학습해 모델을 생성하고, 이렇게 생성된 모델에 새로운 데이터 값이 주어졌을 때 미지의 레이블 값을 예측하는 것입니다.  

**분류**는 다양한 머신러닝 알고리즘으로 구현됩니다.

- Naive Bayes
- Logistic Regression
- **Decision Tree**
- Support Vector Machine
- Nearest Neighbor
- Neural Network
- **Ensemble (앙상블)**

이 중 **앙상블**은 서로 다른(또는 같은) 머신러닝 알고리즘을 **결합**

앙상블의 방식은 매우 다양한데 이 중 기본 알고리즘으로 일반적으로 사용하는 것이 **Decision Tree (결정 트리)**

### 결정 트리 모델의 시각화(Decision Tree Visualization)

**데이터에 있는 규칙을 학습을 통해 자동으로 찾아**내 **트리(Tree) 기반의 분류 규칙**을 생성

따라서 **데이터의 어떤 기준을 바탕으로 규칙을 만들어야 가장 효율적인 분류가 될 것인가**가 알고리즘의 성능을 좌우

규칙은 if/else 문으로 표현  

### **용어 정리**

**루트 노드**: 가장 처음 노드이자 **첫 규칙 노드**

**규칙 노드**: **데이터를 분할하는 규칙 조건**을 나타냄

**리프 노드**: 규칙 노드에 의해 **분류된 분류값**을 나타냄

**서브 트리:** 규칙 노드에 의해 데이터 분할 때마다, 그 규칙 노드를 루트로 하는 새로운 트리가 생성 

             → **서브 트리**

**깊이**: 트리가 나뉜 횟수(?)  
  
### **특징**

- 쉽고 유연하게 적용 가능 → 스케일링, 정규화 등의 사전 가공 영향 적음
- **예측 성능을 향상시키기 위해 복잡한 규칙 구조 필요** (depth 大)
- but, 너무 **복잡한 규칙 구조시 과적합이 발생**해 **예측 성능이 저하**
- 최대한 많은 데이터 세트가 해당 분류에 속하도록 **데이터 셋이 최대한 균일하게 구성**되어야 함  

### **균일(Homogeneous)?**

데이터 셋이 균일하다는 것은 **순수(Pure)**하다는 것을 뜻함

즉, **원소들 간**의 **동질성이 높아야 함**

**균일도가 높다면, 적은 수의 규칙으로 쉽게 원소들을 분류**할 수 있기 때문에, 

결정트리 **학습모델**은 **균일도가 높은 데이터 세트를 먼저 선택**

ex. 30개의 레고 블록 중 ‘형태’ 피처는 동그라미, 네모, 세모, ‘색깔’ 피처는 노랑, 빨강, 파랑이 있다. 노랑 블록의 형태는 모두 동그라미, 빨강, 파랑 블록은 형태가 다 섞여있다고 하자.

학습 모델은 먼저 **‘색깔이 노란색인지’**를 기준으로 규칙 노드를 구성할 것이다. 

그렇다면 **정보의 균일도는 어떻게 측정?**  
  
  
### **정보의 균일도**

- **정보 이득(Information Gain)**
    
    **엔트로피**라는 개념을 기반으로 합니다. 

    **엔트로피** -> 주어진 데이터 집합의 혼잡도
    
    즉, 서로 **다른 값이 섞여** 있으면 **엔트로피가 높고**, **같은 값이 섞여** 있으면 **엔트로피가 낮습니다**. 
    
    **정보 이득 지수**는 1에서 엔트로피 지수를 뺀 값을 사용합니다. 따라서, **(1 - 엔트로피 지수)** 라고 볼 수 있습니다. 이 **정보 이득이 높은 속성을 기준으로 분할**합니다. **(엔트로피가 낮은)**  
    
    
- **지니 계수**
    
    **지니 계수**
    경제학에서 **불평등 지수**를 나타낼 때 사용하는 계수. **0이 가장 평등, 1이 가장 불평등**합니다
    
    결정 트리에서는 **지니 계수가 낮을수록 데이터 균일도가 높은 것**으로 해석합니다. 즉, **지니 계수가 낮은 속성을 기준으로 분할**하게 됩니다.
    
위에서 소개한 내용을 기반으로 했을 때, 결정 트리 모델의 가장 큰 특징은 바로 **정보의 균일도**입니다. 정보의 균일도가 가장 중요하기 때문에, 스케일링이나 정규화 등의 **전처리 작업이 필요 없습니다**. 또한, **알고리즘이 쉽고 직관적**입니다.

반면에, **과적합으로 인해 정확도가 떨어질 수 있습니다.** 그렇기 때문에 차라리 **트리의 크기를 사전에 제한**하는 것이 성능 튜닝에 더 도움이 될 수 있습니다.

### 파라미터 종류  

min_samples_split
- 노드를 분할하기 위한 최소한의 샘플 데이터 수 
    → 최소 n개는 동일한 것이 있어야 규칙 노드를 통해 분할
- default = 2, 작을수록 과적합 가능성 증가
- 과적합 제어. 1로 설정하면 분할되는 노드가 많아져 과적합 가능성 증가

min_samples_leaf
- leaf 노드가 되기 위한 최소한의 샘플 데이터의 수
    → 최소 n개는 되어야 더 분할하지 않고 leaf 노드로 인정
- min_samples_split과 유사하게 과적합을 제어하는 용도 
- 비대칭적인 데이터의 경우 특정 클래스의 데이터가 극도로 작을 수 있으므로 이럴 때는 작게 설정 필요

max_features
- 최적의 분할을 위해 고려할 최대 피처 수. default = None (모든 피처 사용)
- 지정된 값 만큼 무작위로 피처를 선택
- int형 → 대상 피처 개수, float형 → 전체 피처 중 대상 피처 퍼센트
- sqrt → √전체피처개수 만큼 선정 (auto와 동일)
- log → 전체 피처 중 log2(전체피처개수) 선정

max_depth
- 트리 최대 깊이를 규정한다. default = None (완벽하게 결정될 때 까지 또는 min_samples_split보다 작아질 때까지)
- 깊이가 깊어지면 min_samples_split 설정대로 최대 분할해 과적합할 수 있으므로 적절한 값으로 제어 필요함

max_leaf_nodes
- leaf 노드의 최대 개수

### 결정 트리(Decision TREE) 과적합(Overfitting)

make_classification() 함수를 제공  
함수를 이용해 n개의 피쳐가 m가지 유형의 클래스 값을 가지는 데이터셋을 만들 수 있다.  
반환되는 객체는 피쳐 데이터 세트와 클래스 레이블 데이터 세트임  

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("3 Class values with 2 Features Sample data creation")

# 2차원 시각화를 위해서 feature는 2개, 결정값 클래스는 3가지 유형의 classification 샘플 데이터 생성. 
X_features, y_labels = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_classes=3, n_clusters_per_class=1,random_state=0)
# 2개의 feature(n_features)를 3개의 유형(n_classes)으로 분류했다.


# plot 형태로 2개의 feature로 2차원 좌표 시각화, 각 클래스값은 다른 색깔로 표시됨. 
plt.scatter(X_features[:, 0], X_features[:, 1], marker='o', c=y_labels, s=25, cmap='rainbow', edgecolor='k')
```

```python
import numpy as np


# 데이터의 점과 색 뿐이 아니라 boundary를 시각화해서 더 잘보이게 함
# Classifier의 Decision Boundary를 시각화 하는 함수
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # 학습 데이타 scatter plot으로 나타내기
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # 호출 파라미터로 들어온 training 데이타로 model 학습 . 
    model.fit(X, y)
    # meshgrid 형태인 모든 좌표값으로 예측 수행. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# 특정한 트리 생성 제약없는 결정 트리의 Decsion Boundary 시각화.
dt_clf = DecisionTreeClassifier(random_state=156).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```

```python
# min_samples_leaf=6 으로 트리 생성 조건을 제약한 Decision Boundary 시각화
# 제약을 걸은 후의 결과
dt_clf = DecisionTreeClassifier(min_samples_leaf=6, random_state=156).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# features.txt 파일에는 피처 이름 index와 피처명이 공백으로 분리되어 있음. 이를 DataFrame으로 로드.
feature_name_df = pd.read_csv('./features.txt',sep='\s+',
                        header=None,names=['column_index','column_name'])

# 피처명 index를 제거하고, 피처명만 리스트 객체로 생성한 뒤 샘플로 10개만 추출
feature_name = feature_name_df.iloc[:, 1].values.tolist()
print('전체 피처명에서 10개만 추출:', feature_name[:10])
```

중복된 피쳐들이 있기 때문에, 중복된 피쳐를 제거하고 df로 불러올 것

**중복된 피처명을 확인**  
feature_name_df.groupby('column_name')  
을 통해서 column name이 같은 것들을 묶어 볼 수 있음  

```python
feature_dup_df = feature_name_df.groupby('column_name').count()
print(feature_dup_df[feature_dup_df['column_index'] > 1].count())
feature_dup_df[feature_dup_df['column_index'] > 1].head()
```

**원본 데이터에 중복된 Feature 명으로 인하여 신규 버전의 Pandas에서 Duplicate name 에러를 발생.**  
**중복 feature명에 대해서 원본 feature 명에 '_1(또는2)'를 추가로 부여하는 함수인 get_new_feature_name_df() 생성**

이렇게 수정해서 새 df를 반환

```python
def get_new_feature_name_df(old_feature_name_df):
    feature_dup_df = pd.DataFrame(data=old_feature_name_df.groupby('column_name').cumcount(),
                                  columns=['dup_cnt'])
    # dup_cnt를 column으로 갖는 df
    feature_dup_df = feature_dup_df.reset_index()
    # dup_df의 index reset해서 간단한 정수형
    
    new_feature_name_df = pd.merge(old_feature_name_df.reset_index(), feature_dup_df, how='outer')
    #이렇게 old_feature_name_df에 feature_dup_df를 outer merge/join
    new_feature_name_df['column_name'] = new_feature_name_df[['column_name', 'dup_cnt']].apply(lambda x : x[0]+'_'+str(x[1]) 
                                                                                         if x[1] >0 else x[0] ,  axis=1)
    # _ 를 통해서 새 이름을 지어줍니다.
    new_feature_name_df = new_feature_name_df.drop(['index'], axis=1)
    # 이후 index는 drop 
    return new_feature_name_df
```

```python
import pandas as pd

def get_human_dataset( ):
    
    # 각 데이터 파일들은 공백으로 분리되어 있으므로 read_csv에서 공백 문자를 sep으로 할당.
    feature_name_df = pd.read_csv('./features.txt',sep='\s+',
                        header=None,names=['column_index','column_name'])
    
    # 중복된 피처명을 수정하는 get_new_feature_name_df()를 이용, 신규 피처명 DataFrame생성. 
    new_feature_name_df = get_new_feature_name_df(feature_name_df)
    
    # DataFrame에 피처명을 컬럼으로 부여하기 위해 리스트 객체로 다시 변환
    feature_name = new_feature_name_df.iloc[:, 1].values.tolist()
    
    # 학습 피처 데이터 셋과 테스트 피처 데이터을 DataFrame으로 로딩. 컬럼명은 feature_name 적용
    X_train = pd.read_csv('./train/X_train.txt',sep='\s+', names=feature_name )
    X_test = pd.read_csv('./test/X_test.txt',sep='\s+', names=feature_name)
    
    # 학습 레이블과 테스트 레이블 데이터을 DataFrame으로 로딩하고 컬럼명은 action으로 부여
    y_train = pd.read_csv('./train/y_train.txt',sep='\s+',header=None,names=['action'])
    y_test = pd.read_csv('./test/y_test.txt',sep='\s+',header=None,names=['action'])
    
    # 로드된 학습/테스트용 DataFrame을 모두 반환 
    return X_train, X_test, y_train, y_test


X_train, X_test, y_train, y_test = get_human_dataset()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 예제 반복 시 마다 동일한 예측 결과 도출을 위해 random_state 설정
dt_clf = DecisionTreeClassifier(random_state=156)
dt_clf.fit(X_train , y_train)
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test , pred)
print('결정 트리 예측 정확도: {0:.4f}'.format(accuracy))

# DecisionTreeClassifier의 하이퍼 파라미터 추출
print('DecisionTreeClassifier 기본 하이퍼 파라미터:\n', dt_clf.get_params())
```

### GridSearchCV  이용

```python
from sklearn.model_selection import GridSearchCV
# Gridsearch 모델을 통해 최적의 parameter를 찾는 과정입니다.

params = {
    'max_depth' : [ 6, 8 ,10, 12, 16 ,20, 24]
}

# scoring은 accuracy 지표를 활용함
grid_cv = GridSearchCV(dt_clf, param_grid=params, scoring='accuracy', cv=5, verbose=1)

grid_cv.fit(X_train , y_train)
print('GridSearchCV 최고 평균 정확도 수치:{0:.4f}'.format(grid_cv.best_score_))
# 이렇게 돌리고 나면 결과가 나오는데, grid_cv.best_score_에 제일 좋은 결과가 저장되게 됨

print('GridSearchCV 최적 하이퍼 파라미터:', grid_cv.best_params_)
# 그 때의 parameter를 저장하는 grid_cv.best_params_
```

cv_results_ 속성 (depth와 score 담음)을 df로 만들어서 score를 확인할 겁니다.

```python
# GridSearchCV객체의 cv_results_ 속성을 DataFrame으로 생성. 
cv_results_df = pd.DataFrame(grid_cv.cv_results_)

# max_depth 파라미터 값과 그때의 테스트(Evaluation)셋, 학습 데이터 셋의 정확도 수치 추출
cv_results_df[['param_max_depth', 'mean_test_score']]
```

decision tree의 max_depth를 바꾸면서 성능을 평가

```python
max_depths = [ 6, 8 ,10, 12, 16 ,20, 24]
# max_depth 값을 변화 시키면서 그때마다 학습과 테스트 셋에서의 예측 성능 측정
for depth in max_depths:
    dt_clf = DecisionTreeClassifier(max_depth=depth, min_samples_split=16, random_state=156)
    dt_clf.fit(X_train , y_train)
    pred = dt_clf.predict(X_test)
    accuracy = accuracy_score(y_test , pred)
    print('max_depth = {0} 정확도: {1:.4f}'.format(depth , accuracy))
```

쉽게 넣기 위해, dictionary로 선언해 하는 모습  

```python
params = {
    'max_depth' : [ 8 , 12, 16 ,20], 
    'min_samples_split' : [16, 24],
}

grid_cv = GridSearchCV(dt_clf, param_grid=params, scoring='accuracy', cv=5, verbose=1 )
grid_cv.fit(X_train , y_train)
print('GridSearchCV 최고 평균 정확도 수치: {0:.4f}'.format(grid_cv.best_score_))
print('GridSearchCV 최적 하이퍼 파라미터:', grid_cv.best_params_)
```

이렇게 여러 parameter를 준 후, best_estimator_를 통해 그 parameter 조합을 찾을 수 있음

```python
best_df_clf = grid_cv.best_estimator_
pred1 = best_df_clf.predict(X_test)
accuracy = accuracy_score(y_test , pred1)
print('결정 트리 예측 정확도:{0:.4f}'.format(accuracy))
```
