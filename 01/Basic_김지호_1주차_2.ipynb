# DataFrame과 리스트,딕셔너리,넘파이ndarray 상호변환

넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

```python
import numpy as np
#1차원 데이타기반
col_name1=['col1'] #리스트라 칼럼명을 따로 만들어 붙혀줘야함.
list1 = [1, 2, 3]
array1 = np.array(list1) # col1이라는 칼럼명을 지닌 1차원 ndarray 생성
print('array1 shape:', array1.shape ) # shape 출력


# 리스트를 이용해 DataFrame 생성.
df_list1 = pd.DataFrame(list1, columns=col_name1) #데이타프레임-> list1내용으로 만들되 칼럼명을 선언해줄 수 있음.
print('1차원 리스트로 만든 DataFrame:\n', df_list1)


# 넘파이 ndarray를 이용해 DataFrame 생성.
df_array1 = pd.DataFrame(array1, columns=col_name1)#데이타프레임-> array1내용으로 만들되 칼럼명을 선언해줄 수 있음.
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
#2차원 data기반
# 3개의 칼럼명이 필요함.
col_name2=['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape )
df_list2 = pd.DataFrame(list2, columns=col_name2)#데이타프레임-> list2내용으로 만들되 칼럼명을 선언해줄 수 있음.
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)#데이타프레임-> array2내용으로 만들되 칼럼명을 선언해줄 수 있음.
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# Key는 컬럼명으로 매핑, Value는 리스트 형(또는 ndarray)
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

```python
# DataFrame을 values를 이용해 ndarray로 변환
array3 = df_dict.values #values->데이터프레임을 ndarray로 변환해줌.
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape) #array3:dataframe의 행값
print(array3)

# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist() #values로 ndarray로 변환된 값을 tolist() 함수로 호출해서 리스트로 변환
print('df_dict.values.tolist() 타입:', type(list3)) #
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list') #to_dict() 메서드를 호출해 인자로 'list'를 입력하여 값이 리스트형으로 반환된 딕셔너리로의 변환
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

# apply lambda 식으로 데이터 가공

```python
def get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))

#람다를 이용하면 위의 함수를 다음과 같이 간단하게 표현 가능
lambda_square = lambda x : x ** 2
print('3의 제곱은:',lambda_square(3))

a=[1,2,3]
squares = map(lambda x : x**2, a) #map함수를 결합해 lambda에 여러 개의 값을 입력인자로 사용가능
list(squares)
```

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x)) 
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' ) #람다를 이용해 15세 이상이면 Adult 미만이면 Child로 구분하는 새로운 칼럼 생성
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 
                                                                                  'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성. 
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정. 
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
    
```
